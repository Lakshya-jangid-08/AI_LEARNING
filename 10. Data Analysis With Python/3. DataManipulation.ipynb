```python
import pandas as pd
```

```python
df = pd.read_csv('random_people.csv')
```

```python
df.describe()  # statistical summary of numerical columns
```

```python
df.isnull()  #give true , when the value is missing
```

```python
df.isnull().any(axis=0)  #check if any value is missing in a column
# df.isnull().any(axis=1)  #check if any value is missing in a row
```

```python
df.isnull().sum()  # count of missing values in each column
```

```python
df_filled = df.fillna(0)  # fill missing values with 0
```

```python
df['Salary_New'] = df['Salary'].fillna(df['Salary'].mean())  # fill missing values with mean of the column
df
```

```python
# Renaming Columns
df.rename(columns={'Salary_New': 'Salary_Updated'})
df.head()
```

```python
df.dtypes
```

```python
# Change Data Types
# df['Age_INT'] = df['Age'].astype('int')
# This give error bcz of Cannot convert non-finite values (NA or inf) to integer

df['Age_INT'] = df['Age'].fillna(df['Age'].mean()).astype('int')
df
```

```python
# Salary Increment
df['Salary_Increment'] = df['Salary'].apply(lambda x: x*0.1 + x if x<50000 else x*0.05 + x)
df
```

```python
# Data Aggregating & Grouping
# Group by Age with salary
df.groupby('Gender')['Salary_Increment'].sum()
```

```python
df.groupby(['Department', 'Gender'])['Salary_Increment'].sum()
```

```python
# Aggregating functions: sum, mean, median, min, max, count, std, var
grp_agg = df.groupby('Department').agg({'Salary_Increment': ['sum', 'mean', 'max'], 'Age_INT': 'mean'})
grp_agg
```

```python
# Merging DataFrames
df1 = pd.DataFrame({
    "Department" : ["HR", "IT", "Finance", "Marketing"],
    "Location" : ["New York", "San Francisco", "Chicago", "Los Angeles"]
})
df2 = pd.DataFrame({
    "Department" : ["HR", "IT", "Medical", "Marketing"],
    "Budget" : [50000, 60000, 70000, 80000]
})
print(df1)
print()
print(df2)
```

```python
pd.merge(df2, df1, on='Department', how='inner')  # inner, outer, left, right
```

```python
pd.merge(df1, df2, on='Department', how='outer')  # inner, outer, left, right
```
